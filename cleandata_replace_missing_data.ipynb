```python
import pandas as pd 

df = pd.read_excel("excel/Employee.xlsx")
```

```python
df.describe()
```

```python
df
```

```python
#แทนที่ค่าว่างด้วย ค่าเฉลี่ย
select = "Salary"
df[select] = df[select].fillna(df[select].mean())
df
```

```python
#แทนที่ค่าว่างด้วย ค่าที่กำหนดเอง
select = "Salary"
df[select] = df[select].fillna(18000)
df
```

```python
#แทนที่ค่าว่างด้วย ค่าก่อนหน้า
df.fillna(method='pad',inplace=True)
df
```

```python
#แทนที่ค่าว่างด้วย ค่าถัดไป
df.fillna(method='bfill',inplace=True)
df
```
